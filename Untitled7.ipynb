```python
# Instal OpenCV
!pip install opencv-python

# Impor pustaka
from IPython.display import HTML, display, Javascript, clear_output
from google.colab.output import eval_js
from base64 import b64decode, b64encode
from io import BytesIO
from PIL import Image as PILImage
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Unduh cascade
!wget -O haarcascade_frontalface_default.xml https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
!wget -O haarcascade_eye.xml https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_eye.xml
!wget -O haarcascade_smile.xml https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_smile.xml

# Muat classifier
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
smile_cascade = cv2.CascadeClassifier('haarcascade_smile.xml')

# HTML + JavaScript untuk kamera dan tombol
html_code = """
<style>
#video-feed { border:3px solid #333; width:480px; }
button { padding:10px 15px; margin:5px; font-size:16px; cursor:pointer; }
</style>

<h3>ðŸ“¸ Kamera + Deteksi Viola-Jones</h3>
<video id="video-feed" autoplay playsinline></video><br>
<button onclick="captureAndSend()">ðŸŽ¯ Ambil Gambar & Deteksi Wajah</button>

<script>
let video = document.getElementById('video-feed');
let stream = null;

// Buka kamera
navigator.mediaDevices.getUserMedia({video: true})
  .then(s => {
    stream = s;
    video.srcObject = s;
  })
  .catch(err => {
    alert("Gagal mengakses kamera: " + err);
  });

function captureAndSend() {
  // Buat canvas sementara
  const canvas = document.createElement('canvas');
  canvas.width = video.videoWidth;
  canvas.height = video.videoHeight;
  const ctx = canvas.getContext('2d');
  ctx.drawImage(video, 0, 0);

  // Ambil data gambar
  const dataUrl = canvas.toDataURL('image/jpeg', 0.85);

  // Kirim ke Python via google.colab.kernel
  google.colab.kernel.invokeFunction('notebook.capture_frame', [dataUrl], {});
}
</script>
"""

# Fungsi Python yang dipanggil dari JavaScript
def receive_and_process(data_url):
    clear_output(wait=True)
    display(HTML(html_code))  # Render ulang HTML agar tetap ada

    # Decode gambar
    header, encoded = data_url.split(",", 1)
    image_data = b64decode(encoded)
    pil_image = PILImage.open(BytesIO(image_data)).convert('RGB')

    # Konversi ke OpenCV
    img = np.array(pil_image)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Deteksi wajah
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)  # Biru: wajah

        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]

        # Mata
        eyes = eye_cascade.detectMultiScale(roi_gray, scaleFactor=1.1, minNeighbors=8, minSize=(15, 15))
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)  # Hijau

        # Mulut
        smiles = smile_cascade.detectMultiScale(roi_gray, scaleFactor=1.8, minNeighbors=20, minSize=(25, 25))
        for (sx, sy, sw, sh) in smiles:
            cv2.rectangle(roi_color, (sx, sy), (sx + sw, sy + sh), (0, 0, 255), 2)  # Merah

    # Tampilkan hasil
    result_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8, 6))
    plt.imshow(result_rgb)
    plt.axis('off')
    plt.title('âœ… Hasil Deteksi Viola-Jones')
    plt.show()

    print(f"Jumlah wajah terdeteksi: {len(faces)}")
    print(f"Jumlah mata terdeteksi: {len(eyes)}")
    print(f"Jumlah Mulut terdeteksi: {len(smiles)}")

# Daftarkan fungsi ke kernel Colab
from google.colab import output
output.register_callback('notebook.capture_frame', receive_and_process)

# Tampilkan antarmuka
display(HTML(html_code))
```
